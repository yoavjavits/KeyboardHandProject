# Models' performance analysis

We are tyring to train a model such that given a sequence of hand positions, predict the next action - 1 for typing on the keyboard and 0 for not typing. 

The hand positions are represented as a vector of 6 elements, 3 for the x, y, z coordinates of thumb3 and 3 for the x, y, z coordinates of thumb4.
Thumb3 and Thumb4 are calculated relatively to the keyboard's center position, and are normalized using min-max scaling.
It is important to note that the normalization is done for each person separately - we calculate his min-max values and normalize his data according to them.

## First Version

At first, we tried to train a simple LSTM model for our task on our data.
As you can see in the data_analysis notebook, the data is highly imbalanced. We have 6 times more 0s than 1s.
In addition to that, the 0s and 1s are organized in chucks, meaning that we have sequences of 0s and 1s.

There is a detailed analysis of the data, mainly its problems, in the data_analysis notebook.

In the first version we train few LSTM models on the data (after preprocessing it), without any adjustments to the data or to the models.
We chose the model according to **AUC** validation score. 

##### We have two types of training methods:
1. In the first training method, called "many", we train the model in the following way:
- Given a sequence of hand positions, predict for each position if the person is typing or not.

2. In the second training method, called "one", we train the model in the following way:
- Given a sequence of hand positions, predict for the final position if the person is typing or not. The prediction of course is based on the previous positions as well as the final one.

## Second Version

To improve the results we got in the first version I tried to use focal loss and to balance the data.

### Balancing the data
The full analysis of the data can be found in the notebook 'data_analysis'. 
The final chosen process is the following:

For each dataframe (recording):
1. Calculate "heuristic distance".
For each consecutive 1s sequence, calculate the distance between the first and the last position, for thumb3 and thumb4 separately.
The average of the distances is the heuristic distance.

2. Sample the dataframe such that the distance between any two rows is at least the heuristic distance.

<br><br>

### Focal Loss
Focal loss is a loss function that is used to address the class imbalance problem. 
The loss is defined as:
$$FL(p_t) = -\alpha(1 - p_t)^\gamma \log(p_t)$$
where $p_t$ is the predicted probability, $\alpha$ is the balanced parameter and $\gamma$ is a focusing parameter.

As $\alpha$ increases the training will be more focused on correctly classifying positive samples.
As $\gamma$ increases the training will be more focused on "hard samples" - samples where the model is not very confident about the prediction.

The hyperparameters are very sensitive and should be chosen carefully.

#### I have tried three architectures over a set of $\alpha$ and $\gamma$ values.

**First architecture**
Batch size: 32
Learning rate: 0.001
Sequence length: 3
Hidden size: 128
Number of layers: 2

**Second architecture**
Batch size: 32
Learning rate: 0.001
Sequence length:5
Hidden size: 256
Number of layers: 4

**Third architecture**
Batch size: 64

Sequence length: 7
Hidden size: 512
Number of layers: 8

```python
def get_model_params(model_folder):
    model_params_file_path = model_folder + '/model_params.txt'
    
    if not os.path.exists(model_params_file_path):
        return None

    model_params_lines = open(model_params_file_path, 'r').read().split('\n')
    model_type = model_params_lines[0].split(' ')[-1]
    model_seq_len = int(model_params_lines[4].split(' ')[-1])
    model_arch = 1 if model_seq_len == 3 else 2 if model_seq_len == 5 else 3
    model_alpha = float(model_params_lines[8].split(' ')[-1])
    model_gamma = float(model_params_lines[9].split(' ')[-1])
    
    return {'type': model_type, 'arch': model_arch, 'alpha': model_alpha, 'gamma': model_gamma, 'folder': model_folder}
```

```python
def get_model_auc(model_folder):
    model_auc_file_path = model_folder + '/evaluation/results.txt'
    
    if not os.path.exists(model_auc_file_path):
        return None
    
    return open(model_auc_file_path, 'r').read().split('\n')[3].split(' ')[-1]
```

```python
def plot_models_graph(models_dir='./second_version'):
    all_models = [m for m in os.listdir(models_dir) if m.startswith('model')]   
    
    models_params = [get_model_params(models_dir + '/' + m) for m in all_models]
    models_params = [m for m in models_params if m is not None]
    
    models_arch_1 = [m for m in models_params if m['arch'] == 1]
    models_arch_2 = [m for m in models_params if m['arch'] == 2]
    models_arch_3 = [m for m in models_params if m['arch'] == 3]
    
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    
    ax[0].scatter([m['alpha'] for m in models_arch_1], [m['gamma'] for m in models_arch_1], label='Architecture 1')
    ax[0].set_title('Architecture 1')
    ax[0].set_xlabel('Alpha')
    ax[0].set_ylabel('Gamma')
    ax[0].legend()
    
    ax[1].scatter([m['alpha'] for m in models_arch_2], [m['gamma'] for m in models_arch_2], c="red", label='Architecture 2')
    ax[1].set_title('Architecture 2')
    ax[1].set_xlabel('Alpha')
    ax[1].set_ylabel('Gamma')
    ax[1].legend()
    
    ax[2].scatter([m['alpha'] for m in models_arch_3], [m['gamma'] for m in models_arch_3], c="green", label='Architecture 3')
    ax[2].set_title('Architecture 3')
    ax[2].set_xlabel('Alpha')
    ax[2].set_ylabel('Gamma')
    ax[2].legend()
    
    plt.show()
```

```python
plot_models_graph()
```

```python
def plot_models_over_fixed_gamma(ax, gamma, models_dir='./second_version'):
    if gamma not in [2, 4, 6, 8, 10]:
        print('Invalid gamma value')
        return

    all_models = [m for m in os.listdir(models_dir) if m.startswith('model')]

    models_params = [get_model_params(models_dir + '/' + m) for m in all_models]
    models_params = [m for m in models_params if m is not None]
    
    chosen_models = [m for m in models_params if m['gamma'] == gamma]
    auc_values = [float(get_model_auc(m['folder'])) for m in chosen_models]
    
    models_1 = []
    auc_1 = []
    
    models_2 = []
    auc_2 = []
    
    models_3 = []
    auc_3 = []
    
    for m, auc in zip(chosen_models, auc_values):
        if auc is None:
            continue
        
        if m['arch'] == 1:
            models_1.append(m)
            auc_1.append(auc)
        elif m['arch'] == 2:
            models_2.append(m)
            auc_2.append(auc)
        else:
            models_3.append(m)
            auc_3.append(auc)
            
    ax.scatter([m['alpha'] for m in models_1], auc_1, label='Architecture 1')
    ax.scatter([m['alpha'] for m in models_2], auc_2, c='red', label='Architecture 2')
    ax.scatter([m['alpha'] for m in models_3], auc_3, c='green', label='Architecture 3')
    
    ax.set_title(f'AUC values over alpha for gamma={gamma}')
    ax.set_xlabel('Alpha')
    ax.set_ylabel('AUC')
    ax.legend()    
```

```python
def plot_models_over_gammas(models_dir='./second_version'):    
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.tight_layout(pad=5.0)

    for i, gamma in enumerate([2, 4, 6, 8, 10]):
        row = i // 3
        col = i % 3
        plot_models_over_fixed_gamma(axes[row, col], gamma, models_dir)

    plt.show()
```

```python
plot_models_over_gammas()
```

## Best models

```python
def get_top_3_models(models_dir='./second_version'):
    all_models = [m for m in os.listdir(models_dir) if m.startswith('model')]

    models_params = [get_model_params(models_dir + '/' + m) for m in all_models]
    
    # Remove models with missing data
    models_params = [m for m in models_params if m is not None]
    models_params = [m for m in models_params if get_model_auc(m['folder']) is not None]
    
    # Sort models by AUC
    models_params = sorted(models_params, key=lambda x: float(get_model_auc(x['folder'])), reverse=True)
    
    return models_params[:3]
```

```python
def plot_model_results(model_folder):
    img = plt.imread(model_folder + '/evaluation/positive_samples_results_chart.png')
    plt.figure(figsize=(8,6))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    
    img = plt.imread(model_folder + '/evaluation/results_chart.png')
    plt.figure(figsize=(8,6))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

#### Metrics comparison

Accuracy is the number of correct predictions divided by the total number of predictions.

Precision is $\frac{TP}{TP + FP}$, where TP is the number of true positives and FP is the number of false positives. It is important where the cost of false positive is high.

Recall is $\frac{TP}{TP + FN}$, where FN is the number of false negatives. It is important where the cost of false negative is high.

F1 score is the harmonic mean of precision and recall.

AUC is the area under the ROC curve. It is a good metric for imbalanced data. It essentially tells you how well the model can distinguish between the two classes.
