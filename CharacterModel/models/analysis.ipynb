# Models' Analysis

## Reformer Model

Reformer model is a transformer model that uses locality-sensitive hashing (LSH) to reduce the memory complexity of the model. The model is pre-trained on the 'enwik8' dataset by google, and fine-tuned on the 'sentences IMDB' dataset by myself.

The first fine-tuned model was over-fitting the training set - the training loss was decreasing while the validation loss was increasing.
The results of the first fine-tuned model are presented below.

### Architecture
